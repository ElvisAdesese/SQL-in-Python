# Sorting data by the GROUP BY and COUNT KEY WORD

```python
#THE GROUP BY AND COUNT KEY WORD ALOWS YOU TO PERFROM SOME LEVEL STATS ON YOUR DATA. Also note that the count keyword helps you 
#sum the total number of values in field. 
cur.execute('SELECT country_of_birth, COUNT(*) 
            FROM mock_data GROUP BY country_of_birth ORDER BY 
            country_of_birth ASC;')
```

```python
cur.fetchall()
```

# Sorting data by the GROUP BY, HAVING and COUNT KEY WORD


```python
#AGGREGATION
cur.execute('SELECT country_of_birth, COUNT(*) FROM mock_data GROUP BY country_of_birth HAVING COUNT(*)>5 
            ORDER BY country_of_birth ASC;')
```

```python
cur.fetchall()
```

```python
cur.execute('SELECT country_of_birth, COUNT(*) 
            FROM mock_data GROUP BY country_of_birth HAVING COUNT(*)>=10 
            ORDER BY country_of_birth ASC;')
```

```python
cur.fetchall()
```

```python
cur.execute('SELECT country_of_birth, COUNT(*) FROM mock_data 
            GROUP BY country_of_birth HAVING COUNT(*)<=10 
            ORDER BY country_of_birth ASC;')
```

```python
cur.fetchall()
```
